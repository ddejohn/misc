```python
import plotly.graph_objects as go
from math import sin, cos, pi
```

```python
n = 100
theta = [k*2*pi/n for k in range(n+1)]

r_0 = [*map(lambda t: 0.8 + cos(t), theta)]
r_1 = [*map(lambda t: 1.0 + cos(t), theta)]
r_2 = [*map(lambda t: 1.2 + cos(t), theta)]

x_0 = [r*cos(t) for r, t in zip(r_0, theta)]
x_1 = [r*cos(t) for r, t in zip(r_1, theta)]
x_2 = [r*cos(t) for r, t in zip(r_2, theta)]

y_0 = [r*sin(t) for r, t in zip(r_0, theta)]
y_1 = [r*sin(t) for r, t in zip(r_1, theta)]
y_2 = [r*sin(t) for r, t in zip(r_2, theta)]
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=x_0, y=y_0, name="r_0 = 0.8", mode="lines", line_width=4))
fig.add_trace(go.Scatter(x=x_1, y=y_1, name="r_0 = 1.0", mode="lines", line_width=4))
fig.add_trace(go.Scatter(x=x_2, y=y_2, name="r_0 = 1.2", mode="lines", line_width=4))

fig.update_layout(width=800, height=750, xaxis_showgrid=False, yaxis_showgrid=False)
fig.update_xaxes(range=[-0.5, 2.5])
fig.update_yaxes(range=[-1.75, 1.75])
fig.show()
```
