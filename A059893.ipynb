Let $a = \text{A}059893$ and $a(1) = 1$.

$$
a(n) = \begin{cases}
2 \, a \, (n-2^k) & 2^{k+1} \leq n \lt 3 \cdot 2^k \\
1 + a(n-2^k) & 3 \cdot 2^k \leq n \lt 2^{k+2}
\end{cases}
$$

```python
import plotly.graph_objects as go


# A059893|
def a(n):
    return int("1" + bin(n)[3:][::-1], 2)
```

```python
n = 1000
x = [*range(1, n+1)]
y = [a(k) for k in x]

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, name="A059893", mode="markers", marker_size=5))
fig.update_layout(width=900, height=800, xaxis_showgrid=False, yaxis_showgrid=False, yaxis_zeroline=False, xaxis_zeroline=False)
fig.update_yaxes(type="log")
fig.show()
```
