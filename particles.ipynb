```python
import numpy as np
import plotly.graph_objects as go


class Particle:
    def __init__(self, r: list, v: list, dt: float):
        self.r = np.array(r)
        self.v = np.array(v)
        self.dt = dt


    def __str__(self):
        return f"r: {self.r}\nv: {self.v}"


#     @property
#     def direction(self):
#         return np.array([self.dt if 0.0 <= self.r[0] <= 10.0 else -self.dt,
#                          self.dt if 0.0 <= self.r[1] <= 10.0 else -self.dt])
```

```python
p = Particle(r=[1.5, 6.2], v=[-0.84, -0.23], dt=0.05)
x, y = [], []


for _ in range(1000):
    p.v = p.v * np.array([1 if 0.0 <= p.r[0] <= 10.0 else -1,
                          1 if 0.0 <= p.r[1] <= 10.0 else -1]) + np.array([0.0, -10.0]) * p.dt**2
    p.r = p.r + p.v * p.dt
    x.append(p.r[0])
    y.append(p.r[1])


fig = go.Figure()
fig.add_trace(go.Scatter(x=x,
                         y=y,
                         name="particle",
                         mode="markers",
                         marker_size=5))

fig.update_layout(width=900,
                  height=800,
                  xaxis_showgrid=False,
                  yaxis_showgrid=False)
fig.show()
```
